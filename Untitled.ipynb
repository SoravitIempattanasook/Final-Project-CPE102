```python
#นายสรวิศย์ เอี่ยมพัฒนาสุข 5216
import datetime as dt
import matplotlib.patches as mpatches
df['เวลาเข้านอนโดยประมาณ'] = pd.to_datetime(df['เวลาเข้านอนโดยประมาณ'], infer_datetime_format=True, format='%H:%M:%S').dt.time.astype(str)
df = df.sort_values(by = 'เวลาเข้านอนโดยประมาณ')


x = df['เวลาเข้านอนโดยประมาณ']
y = df['เวลานอนรวม']

size = []
colors = []
duplicated_list = list(df.duplicated(['เวลาเข้านอนโดยประมาณ', 'เวลานอนรวม']))

for col_idx,data in df.iterrows():
    if data['จำนวนครั้งในการดื่มเครื่องดื่มที่มีคาเฟอีนในหนึ่งวันโดยเฉลี่ย'] == 0:
        colors.append('blue')
    elif data['จำนวนครั้งในการดื่มเครื่องดื่มที่มีคาเฟอีนในหนึ่งวันโดยเฉลี่ย'] == 1:
        colors.append('yellow')
    else:
        colors.append('red')
for i in duplicated_list:
    if i == True:
        size.append(300)
    else:
        size.append(150)

 
fig, ax = plt.subplots()

plt.rcParams['figure.figsize'] = 15, 6

ax.scatter(x, y, s=size, c=colors, alpha=0.5)

ax.set_xlabel('Bedtime')
ax.set_ylabel('Number of sleep (hours)')

blue_patch = mpatches.Patch(color='blue', label='Number of drinks containing 0 caffeine per day')
yellow_patch = mpatches.Patch(color='yellow', label='Number of drinks containing 1 caffeine per day')
red_patch = mpatches.Patch(color='red', label='Number of drinks containing 2 caffeine per day')
plt.legend(handles=[blue_patch, yellow_patch, red_patch])

ax.grid(True)
```

```python
#นายสรวิศย์ เอี่ยมพัฒนาสุข 5216
Morning=df1[df1.เวลาที่ดื่มเครื่องดื่มที่มีคาเฟอีน.isin(['ช่วงเช้า;', 'ช่วงสาย;'])]
```

```python
#นายสรวิศย์ เอี่ยมพัฒนาสุข 5216
Noon=df1[df1.เวลาที่ดื่มเครื่องดื่มที่มีคาเฟอีน.isin(['ช่วงกลางวัน;'])]
```

```python
#นายสรวิศย์ เอี่ยมพัฒนาสุข 5216
Afternoon=df1[df1.เวลาที่ดื่มเครื่องดื่มที่มีคาเฟอีน.isin(['ช่วงบ่าย;'])]
```

```python
#นายสรวิศย์ เอี่ยมพัฒนาสุข 5216
Evening=df1[df1.เวลาที่ดื่มเครื่องดื่มที่มีคาเฟอีน.isin(['ช่วงเย็น;', 'ช่วงค่ำ;'])]
```

```python
#นายสรวิศย์ เอี่ยมพัฒนาสุข 5216
fig, ax = plt.subplots()

labels = 'Morning', 'Noon', 'Afternoon', 'Evening'
sizes = [int(len(Morning)), int(len(Noon)), int(len(Afternoon)), int(len(Evening))]
colors = ['blue', 'orange', 'green', 'red']
ax.pie(sizes, labels=labels, autopct='%1.2f%%')
```

```python
#นายสรวิศย์ เอี่ยมพัฒนาสุข 5216
def calculate_mean(Df):
    total = 0
    for col_idx,data in Df.iterrows():
        total += float(data['เวลานอนรวม'])
    mean = total/len(Df)
    return mean
```

```python
#นายสรวิศย์ เอี่ยมพัฒนาสุข 5216
fig, ax = plt.subplots()

time = ['Morning', 'Noon', 'Afternoon', 'Evening']
sizes = [calculate_mean(Morning), calculate_mean(Noon), calculate_mean(Afternoon), calculate_mean(Evening)]
bar_colors = ['blue', 'orange', 'green', 'red']

ax.barh(time, sizes, color=bar_colors)

ax.set_xlabel('Number of hours of sleep')
ax.set_ylabel('Drinking time')
ax.set_title('Graph showing hours of sleep and time spent drinking caffeinated beverages')

plt.rcParams['figure.figsize'] = 8, 6

plt.show()
```
